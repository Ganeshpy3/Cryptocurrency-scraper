```python
from bs4 import BeautifulSoup
import  json
import requests
import time
import pandas as pd
import datetime
```

```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
from pprint import pprint
```

```python
scope = ["https://spreadsheets.google.com/feeds",'https://www.googleapis.com/auth/spreadsheets',"https://www.googleapis.com/auth/drive.file","https://www.googleapis.com/auth/drive"]
creds = ServiceAccountCredentials.from_json_keyfile_name("credits.json", scope)

client = gspread.authorize(creds)
```

```python
crypto=["loopring","ethereum","bitcoin"]
```

```python
def show_coins(crypto):
    for i in crypto:
        URL="https://coinmarketcap.com/currencies/"+i+"/"
        HEADERS = ({'User-Agent':
                'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
            'Accept-Language': 'en-US'})
        webpage = requests.get(URL,headers=HEADERS)
        state_soup = BeautifulSoup(webpage.content, "lxml")
        price=state_soup.findAll("div",{"class":"priceValue"})
        percent=state_soup.findAll("span",{"class":"sc-15yy2pl-0 feeyND"})
        sheet= client.open(i).sheet1
        for i in price:
            pric=i.text.split(" ")[0]
            pric=float("".join(pric[1:].split(",")))
            break
    
        text=str(percent)
        if(text.find("icon-Caret-down")<0):
            colour="Green"
            
            percent=state_soup.findAll("span",{"class":"sc-15yy2pl-0 gEePkg"})
            for i in percent:
                per=i.text.split(" ")[0]
                break
            
        else:
            colour="RED"
            percent=state_soup.findAll("span",{"class":"sc-15yy2pl-0 feeyND"})
            for i in percent:
                per=i.text.split(" ")[0]
                break
        now=datetime.datetime.now()
            
        data=[now.strftime("%d-%m-%y"),now.strftime("%H:%M:%S"),pric,colour,per]
        sheet.append_row(data)
```

```python
df=None
i=0
while True:
    print("Done")
    

    show_coins(crypto)
    i=i+1
    time.sleep(10)
    if(i>1000):
        break
        
    
```
